```python
from firebase import firebase

# התחברות ל-Firebase Realtime Database עם הקישור שלך
FBconn = firebase.FirebaseApplication('https://cloudtut5-ef56b-default-rtdb.firebaseio.com/', None)

# פונקציות עזר
def add_single_word(word):
    """הוספת מילה בודדת למסד הנתונים"""
    data_to_upload = {
        'count': 1
    }
    result = FBconn.post('/words/' + word, data_to_upload)
    print(f"'{word}' added with count 1.")

def add_text_for_analysis(text):
    """הוספת טקסט ופיצול למילים"""
    words = text.split()
    for word in words:
        # בדיקה אם המילה כבר קיימת במסד הנתונים
        result = FBconn.get('/words/' + word, None)
        if result and 'count' in result:
            current_count = result['count']
            FBconn.put('/words/' + word, 'count', current_count + 1)
        else:
            data_to_upload = {'count': 1}
            FBconn.post('/words/' + word, data_to_upload)
    print(f"Text analyzed and words updated.")

def update_word_count(word, count):
    """עדכון ספירה של מילה"""
    data_to_upload = {'count': count}
    result = FBconn.put('/words/' + word, data_to_upload)
    print(f"'{word}' count updated to {count}.")

def delete_word(word):
    """מחיקת מילה"""
    result = FBconn.delete('/words/' + word, None)
    print(f"'{word}' deleted.")

def view_all_words():
    """הצגת כל המילים"""
    result = FBconn.get('/words', None)
    if result:
        print("Words in database:")
        for word, word_data in result.items():
            # Calculate the total count for each word
            total_count = sum(data['count'] for data in word_data.values())
            print(f"{word}: {total_count}")
    else:
        print("No words in the database.")

# תפריט ראשי
def menu():
    while True:
        print("\nCommon Words Counter Menu:")
        print("1. Add single word")
        print("2. Add text for analysis")
        print("3. Update word count")
        print("4. Delete word")
        print("5. View all words")
        print("6. Exit")

        choice = input("Select an option (1-6): ")
        if choice == '1':
            word = input("Enter a single word: ")
            add_single_word(word)
        elif choice == '2':
            text = input("Enter text for analysis: ")
            add_text_for_analysis(text)
        elif choice == '3':
            word = input("Enter the word to update: ")
            count = int(input("Enter the new count: "))
            update_word_count(word, count)
        elif choice == '4':
            word = input("Enter the word to delete: ")
            delete_word(word)
        elif choice == '5':
            view_all_words()
        elif choice == '6':
            print("Exiting...")
            break
        else:
            print("Invalid option. Please try again.")

# הרצת התפריט
menu()
```
